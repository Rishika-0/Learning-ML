```python
df.drop(columns=['deck']) #since mostly it was empty and was making our model weak(eliminating null values)
```

```python
df.drop(columns=['deck'],inplace=True)
```

```python
df
```

```python
df['age'].value_counts()
```

```python
df['age'].fillna(df['age'])
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)      #replacing with mean is done to eliminate null values 
```

```python
df['age'].value_counts()
```

```python
df['embarked'].value_counts()
```

```python
df['embarked'].fillna('S',inplace=True)    #eliminating null values by replacing with mode(highest value)
```

```python
df.drop(columns=['embark_town'],inplace=True)     #since embarked and embark_town was same
```

```python
df
```

```python
df.info()
```

```python
df['survived']=df['survived'].astype('category')
df['pclass']=df['pclass'].astype('category')
df['sex']=df['sex'].astype('category')
df['age']=df['age'].astype('int')
df['embarked']=df['embarked'].astype('category')
```
